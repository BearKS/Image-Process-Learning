```python
import numpy as np
import cv2
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot as plt
```

```python
#Lab7.1
fig = plt.figure(figsize=(18,18))
img = cv2.imread('Grid_Image.JPG')
img2 = cv2.imread('image.jpeg')
# Define resize factor
Reduce_factors = [2,4,7,10,15]
inter_methods = [cv2.INTER_NEAREST,cv2.INTER_LINEAR,cv2.INTER_CUBIC,cv2.INTER_AREA]
title = ["INTER_NEAREST","INTER_LINEAR","INTER_CUBIC","INTER_AREA"]

for i in range(5):
    Scale_factors = 1/Reduce_factors[i]
    for j in range(4):
        img_resize = cv2.resize(img, None, fx = Scale_factors, fy = Scale_factors, interpolation=inter_methods[j])
        fig.add_subplot(5,4,j+(i*4)+1)
        plt.imshow(cv2.cvtColor(img_resize,cv2.COLOR_BGR2RGB))
        plt.title(title[j] + "(" + str(Reduce_factors[i]) + ")")
        
plt.show()
```
